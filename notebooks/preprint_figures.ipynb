# Preprint Figure Generation

Generates all 6 figures for the VFD Crystallisation preprint.
All outputs are deterministic (seeded RNG) and publication-quality.

```python
import sys
sys.path.insert(0, '../src')

import matplotlib
matplotlib.rcParams.update({
    'font.size': 11,
    'font.family': 'serif',
    'axes.labelsize': 11,
    'axes.titlesize': 12,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'legend.fontsize': 9,
    'figure.dpi': 150,
})

from vfd.crystallisation.figures import (
    plot_functional_descent,
    plot_mode_evolution,
    plot_decoherence_comparison,
    plot_repeatability,
    plot_basin_selection,
    plot_transition_scaling,
    generate_all_figures,
)

OUTPUT_DIR = '../figures'
```

## Figure 1 — Functional Descent

$\mathcal{F}(t)$ monotonically decreases under gradient flow, with decomposition into $R$, $E$, $Q$.

```python
fig1 = plot_functional_descent(f'{OUTPUT_DIR}/fig1_functional_descent.png')
fig1.show()
```

## Figure 2 — Mode Evolution

Spectral coefficients $|c_i(t)|$ and phases showing structured reweighting.

```python
fig2 = plot_mode_evolution(f'{OUTPUT_DIR}/fig2_mode_evolution.png')
fig2.show()
```

## Figure 3 — Decoherence Comparison

Same initial state: crystallisation vs standard decoherence. Outcome distributions and transition times.

```python
fig3 = plot_decoherence_comparison(f'{OUTPUT_DIR}/fig3_decoherence_comparison.png')
fig3.show()
```

## Figure 4 — Repeatability

30 independent experiments: crystallisation gives near-zero entropy (deterministic), stochastic collapse gives Born-rule spread.

```python
fig4 = plot_repeatability(f'{OUTPUT_DIR}/fig4_repeatability.png')
fig4.show()
```

## Figure 5 — Basin Selection

25 initial conditions converge to specific basins in a multi-attractor landscape.

```python
fig5 = plot_basin_selection(f'{OUTPUT_DIR}/fig5_basin_selection.png')
fig5.show()
```

## Figure 6 — Transition-Time Scaling

$\tau_{\mathrm{crys}}$ vs constraint scale and residual magnitude.

```python
fig6 = plot_transition_scaling(f'{OUTPUT_DIR}/fig6_transition_scaling.png')
fig6.show()
```
